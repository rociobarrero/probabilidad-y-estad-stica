###Crear un índice

```python
# Vamos a crear una definición para clasificar (bueno, regular o malo) libros de acuerdo al rating

def rate(rating):
    if rating >= 4:
        return "Bueno"
    elif rating >= 3:
        return "Regular"
    else:
        return "Malo"
```

El método `.apply()` es útil cuando se requiere aplicar una función al conjunto de datos.

```python
# Podemos crear una columna que llamaremos Class

libros_data2["class"] = libros_data2["rating"].apply(rate)
libros_data2[["title", "rating", "name", "num_reviews", "genre", "class"]]
```
