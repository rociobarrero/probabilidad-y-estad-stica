```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
libros.info()
```

```python
from pandas.core.groupby.generic import ScalarResult
import random 
sales = [ ]
for i in range(0, 1539):
  x = round(random.uniform(10.5, 75.5), 3)
  sales.append(x)
libros["sales"] = sales
libros
```

¿Cuáles fueron los 3 libros con menores ventas pero mayor rating?

```python
rating = libros.sort_values("rating", ascending=False).head(3)
rating
```

```python
sales = libros.sort_values("sales", ascending = True).head(3)
sales 
```

```python
rasa = pd.concat([sales, rating], axis = 0)
rasa = sales.append([rating])
rasa = pd.concat([sales, rating], axis = 0)
rasa
```

Los tres libros con menores ventas son: "The Wheel of Darkness", "Next", "The Roanoke Girls" y los libros con mayor rating son "Words of Radiance", "Heartstopper: Volume Four", "The Way of Kings"

Con los datos arrojados, podemos observar que de los 1539 libros que hay en la lista,  el promedio del rating es de 4.037, este ratio se posiciona en un nivel estadar, es decir, por encima de la media entre el rating más y menos elevado. 
Por otro lado, el mayor rating de la base de datos es de 4.75, mientras que el menor es de 2.78.
En cuanto al quartil 1, podemos decir que el 25% tienen un rating menor o igual a 3.905, lo que nos indica que pocos son los libros por debajo del rating promedio. 
En cuanto al quartil 2, podemos observar que el 50% de los datos tienen un rating menor o igual a 4.05, ya que la mitad tiene un rating mayor al promedio.
Finalmente, el quartil 3, nos indica que el 75% de libros tienen un rating menor o igual a 4.18, lo que quiere decir que de la lista, tres cuartas partes de los libros tienen un buen posicionamiento, ya que su rating esta por encima del  promedio.
Como conclusion podemos decir que esta lista tiene libros muy buenos ya que la mayoría tiene buen rating y la comparación que hacemos no varia mucho.
